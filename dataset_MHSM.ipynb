```python
import kagglehub

import os
import pandas as pd
```

```python
# https://www.kaggle.com/datasets/infamouscoder/mental-health-social-media/data



# Download latest version
path = kagglehub.dataset_download("infamouscoder/mental-health-social-media")

print("Path to dataset files:", path)
```

```python
# Find the CSV file within the downloaded directory
for filename in os.listdir(path):
    if filename.endswith(".csv"):
        csv_file_path = os.path.join(path, filename)
        break  # Stop after finding the first CSV file

# Read the CSV file
data = pd.read_csv(csv_file_path)
data.head()
```

```python
data = data[['post_text', 'label']]
data.columns = ['original_text', 'label']
data.head()
```

```python
(data['label'] == 1).sum() # 10,000 per label, balanced
```

```python
len(data)
```

```python
!pip install emoji
import re
import emoji

# Function to clean text
def clean_text(text):
    text = text.lower()  # Convert to lowercase
    text = re.sub(r"http\S+|www\S+", "", text)  # Remove URLs
    text = re.sub(r"@\w+", "", text)  # Remove mentions (@user)
    text = re.sub(r"[^\w\s]", "", text)  # Remove punctuation & special chars
    text = re.sub(r"\s+", " ", text).strip()  # Remove extra spaces
    text = emoji.replace_emoji(text, replace="")  # Remove emojis
    return text
```

```python

data['text'] = data['original_text'].apply(clean_text)
```

```python
# save df to drive
data.to_csv('/content/drive/MyDrive/Colab Notebooks/data_MHSM_20K.csv', index=False)
```

```python
# sample 500 from each label
data_1 = data[data['label'] == 1].sample(500)
data_0 = data[data['label'] == 0].sample(500)

# save to new df and shuffle
data_1000 = pd.concat([data_1, data_0]).sample(frac=1).reset_index(drop=True)
data_1000.head()
```

```python
# save new df to drive
data_1000.to_csv('/content/drive/MyDrive/Colab Notebooks/data_MHSM_1000.csv', index=False)
```
