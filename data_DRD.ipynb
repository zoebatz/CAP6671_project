dataset to use for fine tuning BERT models (~7000 examples)

```python
import kagglehub
import pandas as pd
import os

# Download latest version
path = kagglehub.dataset_download("infamouscoder/depression-reddit-cleaned")

# function to read single csv file from directory
def read_single_csv(path):
  for filename in os.listdir(path):
      if filename.endswith(".csv"):
          csv_file_path = os.path.join(path, filename)
          break  # Stop after finding the first CSV file

  return pd.read_csv(csv_file_path)

data = read_single_csv(path)
data.head()
```

```python
# rename columns
data.rename(columns={'clean_text': 'text', 'is_depression': 'label'}, inplace=True)
# shuffle data
data = data.sample(frac=1).reset_index(drop=True)

data.head()
```
