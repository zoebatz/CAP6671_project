```python

import os
import pandas as pd
```

```python
path = '/content/drive/MyDrive/MDDL/positive/data/tweet'
```

```python
import json

# List to store tweet data
tweets_data = []

# Loop through all JSON files in the folder
for filename in os.listdir(path):
    if filename.endswith(".json"):  # Ensure only JSON files are read
        file_path = os.path.join(path, filename)

        with open(file_path, "r", encoding="utf-8") as file:
            tweet = json.load(file)  # Load JSON content

            # Extract relevant information (modify based on JSON structure)
            tweets_data.append({
                "id": tweet.get("id", None),
                "text": tweet.get("text", None),
                "user": tweet.get("user", {}).get("screen_name", None),
                "created_at": tweet.get("created_at", None),
                "lang": tweet.get("lang", None)
            })

# Convert list of tweets into a DataFrame
df_pos = pd.DataFrame(tweets_data)

# Display first few rows
print(df_pos.head())
```

```python
path = '/content/drive/MyDrive/MDDL/negative/data/tweet'

# List to store tweet data
tweets_data = []

# Loop through all JSON files in the folder
for filename in os.listdir(path):
    if filename.endswith(".json"):  # Ensure only JSON files are read
        file_path = os.path.join(path, filename)

        with open(file_path, "r", encoding="utf-8") as file:
            tweet = json.load(file)  # Load JSON content

            # Extract relevant information (modify based on JSON structure)
            tweets_data.append({
                "id": tweet.get("id", None),
                "text": tweet.get("text", None),
                "user": tweet.get("user", {}).get("screen_name", None),
                "created_at": tweet.get("created_at", None),
                "lang": tweet.get("lang", None)
            })

# Convert list of tweets into a DataFrame
df_neg = pd.DataFrame(tweets_data)

# Display first few rows
print(df_neg.head())
```

```python
!pip install emoji
import re
import emoji

# Function to clean text
def clean_text(text):
    text = text.lower()  # Convert to lowercase
    text = re.sub(r"http\S+|www\S+", "", text)  # Remove URLs
    text = re.sub(r"@\w+", "", text)  # Remove mentions (@user)
    text = re.sub(r"[^\w\s]", "", text)  # Remove punctuation & special chars
    text = re.sub(r"\s+", " ", text).strip()  # Remove extra spaces
    text = emoji.replace_emoji(text, replace="")  # Remove emojis
    return text
```

```python
#df_pos.loc[:, 'text'] = df_pos['text'].apply(clean_text)
#df_neg.loc[:, 'text'] = df_neg['text'].apply(clean_text)

df_pos['clean_text'] = df_pos['text'].apply(clean_text)
df_neg['clean_text'] = df_neg['text'].apply(clean_text)
```

```python
df_neg.head()
```

```python
df_pos = df_pos.rename(columns={'text': 'original_text'})
df_neg = df_neg.rename(columns={'text': 'original_text'})

df_pos = df_pos.rename(columns={'clean_text': 'text'})
df_neg = df_neg.rename(columns={'clean_text': 'text'})
```

```python
# get pos and neg data from drive
path = '/content/drive/MyDrive/Colab Notebooks/data_MDDL_pos_clean.csv'
df_pos = pd.read_csv(path)

path = '/content/drive/MyDrive/Colab Notebooks/data_MDDL_neg_clean.csv'
df_neg = pd.read_csv(path)
```

```python
# 2,000 samples from each label
df_pos = df_pos.sample(n=2000, random_state=42)
df_neg = df_neg.sample(n=2000, random_state=42)

```

```python
# Combine DataFrames (stack rows)
df_combined = pd.concat([df_neg, df_pos], ignore_index=True)
```
